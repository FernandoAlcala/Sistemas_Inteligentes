```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split

# 1. Cargar el dataset
dataset = np.loadtxt(r"C:\Users\popez\OneDrive\Documentos\Sistemas inteligentes\TensorFlow\pima-indians-diabetes.csv", delimiter=",")
X = dataset[:, 0:8]  # Features (Entradas)
Y = dataset[:, 8]    # Labels (Salidas)

# 2. Escalar los datos
scaler = StandardScaler()
X = scaler.fit_transform(X)

# 3. Dividir en entrenamiento (80%) y prueba (20%)
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=7)

# 4. Construir la red neuronal
model = Sequential()
model.add(Dense(32, input_dim=8, activation='relu'))  # Capa oculta 1 (32 neuronas)
model.add(Dense(16, activation='relu'))  # Capa oculta 2 (16 neuronas)
model.add(Dense(8, activation='relu'))  # Capa oculta 3 (8 neuronas)
model.add(Dense(1, activation='sigmoid'))  # Capa de salida (1 neurona con activación sigmoide)

# 5. Compilar el modelo
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

# 6. Entrenar el modelo con los datos de entrenamiento
model.fit(X_train, Y_train, epochs=300, batch_size=32)

# 7. Evaluar el modelo en datos de entrenamiento
train_scores = model.evaluate(X_train, Y_train)
print("\nExactitud en datos de entrenamiento: %.2f%%" % (train_scores[1] * 100))

# 8. Evaluar el modelo en datos de prueba
test_scores = model.evaluate(X_test, Y_test)
print("Exactitud en datos de prueba: %.2f%%" % (test_scores[1] * 100))

# 9. Hacer predicciones con los datos de prueba
predictions = model.predict(X_test)
rounded = [round(x[0]) for x in predictions]
print("Predicciones en datos de prueba:", rounded)
```
