```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import numpy as np

np.random.seed(7)

dataset = np.loadtxt(r"C:\Users\popez\OneDrive\Documentos\Sistemas inteligentes\TensorFlow\pima-indians-diabetes.csv", delimiter=",")
X = dataset[:, 0:8] # Input Features
Y = dataset[:, 8] # Output Features

model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu')) # First Hidden Layer
model.add(Dense(8, activation='relu')) # Second Hidden Layer
model.add(Dense(1, activation='sigmoid')) # Output Layer
          
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])

model.fit(X, Y, epochs=150, batch_size=10)

scores = model.evaluate(X, Y)
print("\n%s: %.2f%%" % (model.metrics_names[1], scores[1]*100))
predictions = model.predict(X)
rounded = [round(x[0]) for x in predictions]
print(rounded)
```
